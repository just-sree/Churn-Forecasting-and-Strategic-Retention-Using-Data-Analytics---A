```python
# Preliminary Analysis Function
# Enhanced EDA for Churn Prediction and Telco Customer Churn Datasets

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
churn_pred_df = pd.read_csv("hf://datasets/scikit-learn/churn-prediction/dataset.csv")
telco_churn_df = pd.read_csv("hf://datasets/aai510-group1/telco-customer-churn/train.csv")

# Enhanced Visualization Function
def enhanced_eda(df, dataset_name):
    print(f"\n--- Enhanced Preliminary Analysis for {dataset_name} ---\n")

    # Churn distribution
    plt.figure(figsize=(8, 5))
    sns.countplot(x='Churn', data=df, palette='viridis')
    plt.title(f'{dataset_name} - Churn Distribution', fontsize=15)
    plt.xlabel('Churn', fontsize=12)
    plt.ylabel('Number of Customers', fontsize=12)
    plt.grid(axis='y', linestyle='--')
    plt.show()

    # Correlation heatmap for numerical variables
    numeric_df = df.select_dtypes(include=['float64', 'int64'])
    plt.figure(figsize=(16, 10))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title(f'{dataset_name} - Correlation Heatmap', fontsize=16)
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.show()

    # Detailed churn analysis by categorical features
    categorical_features = ['Contract', 'PaymentMethod', 'InternetService']
    for feature in categorical_features:
        if feature in df.columns:
            plt.figure(figsize=(10, 6))
            sns.countplot(x=feature, hue='Churn', data=df, palette='coolwarm')
            plt.title(f'{dataset_name} - Churn by {feature}', fontsize=15)
            plt.xlabel(feature, fontsize=12)
            plt.ylabel('Customer Count', fontsize=12)
            plt.xticks(rotation=30)
            plt.grid(axis='y', linestyle='--')
            plt.legend(title='Churn', loc='upper right')
            plt.show()

# Execute enhanced visualization for both datasets
enhanced_eda(churn_pred_df, "Churn Prediction Dataset")
enhanced_eda(telco_churn_df, "Telco Customer Churn Dataset")
```
