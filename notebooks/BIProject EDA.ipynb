# Data Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

splits = {'train': 'train.csv', 'validation': 'validation.csv', 'test': 'test.csv'}
telco_churn_df = pd.read_csv("hf://datasets/aai510-group1/telco-customer-churn/" + splits["train"]) 
churn_pred_df = pd.read_csv("hf://datasets/scikit-learn/churn-prediction/dataset.csv")
churn_pred_df.to_csv("IBM_churn_pred_Dataset.csv", index=False)
telco_churn_df.to_csv("Telco_churn_with_reason_Dataset.csv", index=False)
```

# EDA

```python
# Describe churn_pred_df
print("Churn Prediction Dataset Description:")
print(churn_pred_df.describe(include='all'))
print("\nChurn Prediction Dataset Info:")
print(churn_pred_df.info())
print("\nChurn Prediction Dataset Null Values:")
print(churn_pred_df.isnull().sum())
```

```python
# Describe telco_churn_df
print("\nTelco Customer Churn Dataset Description:")
print(telco_churn_df.describe(include='all'))
print("\nTelco Customer Churn Dataset Info:")
print(telco_churn_df.info())
print("\nTelco Customer Churn Dataset Null Values:")
print(telco_churn_df.isnull().sum())
```

## Data Cleaning and Type conversion

```python
churn_pred_df['TotalCharges'] = pd.to_numeric(churn_pred_df['TotalCharges'], errors='coerce')
churn_pred_df['TotalCharges'].fillna(churn_pred_df['TotalCharges'].median(), inplace=True)

# Address missing values explicitly
telco_churn_df['Churn Reason'].fillna('Not Provided', inplace=True)
telco_churn_df['Offer'].fillna('No Offer', inplace=True)
telco_churn_df['Internet Type'].fillna('Unknown', inplace=True)
```
